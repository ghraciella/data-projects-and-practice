The diagram below shows a model of the business processes covered in the case study. It represents the business processes of a freight forwarding company, in which up to three smaller shipments from suppliers are consolidated and in turn shipped together to customers. The business process is structured into incoming and outgoing transport legs, which jointly aim at ensuring that freight is timely delivered to customers.

<img src = "workflow_UML.png">


Each of the transport legs involves the following physical transport services:
• RCS: Check in freight at departure airline. Shipment is checked in and a receipt is produced at departure airport.
• DEP: Confirm goods on board. Aircraft has departed with shipment on board.
• RCF: Accept freight at arrival airline. Shipment is checked in according to the documents and stored at arrival warehouse.
• DLV: Deliver freight. Receipt of shipment was signed at destination airport.

A transport leg may involve multiple segments (e.g., transfers to other flights or airlines). In those cases, activity RCF loops back to DEP (indicated by the “loop-back” arrow in Figure 1). In the case study, the number of segments per leg ranges from one to four.

The transport services are denoted by three-letter acronyms according to the Cargo 2000 industry standard. Cargo 2000 is an initiative of IATA, the International Air Transport Association (Cargo 2000 has been re-branded as Cargo iQ in 2016). It aims at delivering a new quality management system for the air cargo industry. Cargo 2000 allows for unprecedented transparency in the supply chain. Stakeholders involved in the transport process can share agreed Cargo 2000 messages, comprising transport planning, replanning and service completion events. Cargo 2000 is based on the following key principles: (1) Every shipment gets a plan (called a route map) describing predefined monitoring events. (2) Every service used during shipment is assigned a predefined milestone with a planned time of achievement. (3) Stakeholders receive alerts when a milestone has failed and notifications upon milestone completion, which include the effective time the milestone has been achieved.

<b>The case study data comprises tracking and tracing events from a forwarding company’s Cargo 2000 system for a period of five months. From those Cargo 2000 messages, we reconstructed execution traces of 3,942 actual business process instances, comprising 7,932 transport legs and 56,082 service invocations. Each execution trace includes planned and effective durations (in minutes) for each of the services of the business process (introduced in Section II), as well as airport codes for the DEP (“departure”) and RCF (“arrival”) services. Due to the fact that handling of transport documents along the business process differs based on whether the documents are paper-based or electronic, we focus on the flow of physical goods, as our data set did not allow us to discern the different document types.</b>

The reconstruction process involved data sanitation and anonymization. We filtered overlapping and incomplete Cargo 2000 messages, removed canceled transports (i.e., deleted route maps), sanitized for exceptions from the C2K system (such as events occurring before route map creation) and homogenized the way information was represented in different message types. Finally, due to confidentiality reasons, message fields which might exhibit business critical or customer-related data (such as airway bill numbers, flight numbers and airport codes) have been eliminated or masked. <b> source: S-CUBE</b>

```python
def descriptive_statistic(name):
    name_header = list(name)
    leg = name.convert_objects(convert_numeric=True)
    not_needed = ["i1_dep_1_place", "i1_rcf_1_place", "i1_rcf_2_place", "i1_rcf_3_place", "i2_rcf_1_place", "i2_rcf_2_place",\
                 "i2_rcf_3_place","i3_rcf_1_place", "i3_rcf_2_place", "i3_rcf_3_place", "i1_dep_2_place","i1_dep_3_place",\
                 "i2_dep_1_place","i2_dep_2_place","i2_dep_3_place","i3_dep_1_place","i3_dep_2_place","i3_dep_3_place", \
                  "i2_legid", "i3_legid"]
    for each in not_needed:
        if each in name_header:
            name_header.remove(each)
            leg = leg.drop(each, axis = 1)
        else: 
            continue
    if "i1_legid" in name_header:
        return leg.describe().drop(name_header[0], axis = 1)
    else:
        return leg.describe()

def segment_histograms(name):
    cols = list(name)
    name = name.convert_objects(convert_numeric=True)
    not_needed = ["i1_dep_1_place", "i1_rcf_1_place", "i1_rcf_2_place", "i1_rcf_3_place", "i2_rcf_1_place", "i2_rcf_2_place",\
                 "i2_rcf_3_place","i3_rcf_1_place", "i3_rcf_2_place", "i3_rcf_3_place", "i1_dep_2_place","i1_dep_3_place",\
                 "i2_dep_1_place","i2_dep_2_place","i2_dep_3_place","i3_dep_1_place","i3_dep_2_place","i3_dep_3_place", \
                 "i2_legid", "i3_legid"]
    for each in not_needed:
        if each in cols:
            cols.remove(each)
            name = name.drop(each, axis = 1)
        else: 
            continue
    if "i1_legid" in cols:
        segments = name.drop(cols[0], axis = 1).reset_index(drop = True)
    else:
        segments = name
    hist_Seg = segments.hist(figsize = (20,15), bins=100)
    plt.show()
def correlation_description(name):
    cols = list(name)
    name = name.convert_objects(convert_numeric=True)
    not_needed = ["i1_dep_1_place", "i1_rcf_1_place", "i1_rcf_2_place", "i1_rcf_3_place", "i2_rcf_1_place", "i2_rcf_2_place",\
                 "i2_rcf_3_place","i3_rcf_1_place", "i3_rcf_2_place", "i3_rcf_3_place", "i1_dep_2_place","i1_dep_3_place",\
                 "i2_dep_1_place","i2_dep_2_place","i2_dep_3_place","i3_dep_1_place","i3_dep_2_place","i3_dep_3_place", \
                 "i2_legid", "i3_legid"]
    for each in not_needed:
        if each in cols:
            cols.remove(each)
            name = name.drop(each, axis = 1)
        else: 
            continue
    if "i1_legid" in cols:
        name2 = name.drop(cols[0], axis = 1).reset_index(drop = True)
    else: 
        name2 = name
    correlate = name2.corr()
    f, ax = plt.subplots(figsize=(12,12))
    ax = sns.heatmap(correlate, square = True,ax=ax)
    ax.set_title('Correlation Matrix for the Segment')
    plt.show()
    MatrixBool = np.tril(np.ones(correlate.shape)).astype(np.bool)
    for index in range(0,correlate.shape[0]):
        MatrixBool[index][index] = False
    correlations = correlate.where(MatrixBool)
    print(correlations.stack().nlargest(3))
def scatter_plotter(name):
    cols = list(name)
    name = name.convert_objects(convert_numeric=True)
    
    not_needed = ["i1_dep_1_place", "i1_rcf_1_place", "i1_rcf_2_place", "i1_rcf_3_place", "i2_rcf_1_place", "i2_rcf_2_place",\
                 "i2_rcf_3_place","i3_rcf_1_place", "i3_rcf_2_place", "i3_rcf_3_place", "i1_dep_2_place","i1_dep_3_place",\
                 "i2_dep_1_place","i2_dep_2_place","i2_dep_3_place","i3_dep_1_place","i3_dep_2_place","i3_dep_3_place", \
                 "i2_legid", "i3_legid"]
    for each in not_needed:
        if each in cols:
            cols.remove(each)
            name = name.drop(each, axis = 1)
        else: 
            continue
    if "i1_legid" in cols:
        name2 = name.drop(cols[0], axis = 1).reset_index(drop = True)
    else:
        name2 = name
    g3 = sns.PairGrid(name2, palette='RdBu_r')
    g3.map(plt.scatter, alpha=0.8)
    g3.add_legend()
```

<h3 align="center"><font color="red"> Logistic Regression Analysis on Segment 1, Leg 1 (Supervised Learning)</font></h3>

<p>To start with, we want to run a line graph on the entire leg1 to see how far the planed time aligned with the executed time in the over time. We intend to build a <b>Logistic Regression model</b> to predict the case of Late delivery and On-time delivery of a particular transport leg. Here we have used leg1 as a case study and we shall generalize this concept for other transport legs since the legs are independent variables.</p>

<p>To create the desired Logistic Regression, we calculated the total planned time and total execution time for each customer(occurrence) and then generated a binomial distribution that says if the proposed time is met or not.</p>

<p>Out of the 3945 occurrences we took 80 percent of the data set as our Training Data Set and used our Logistic Regression Model to predict the remaining 20 percent which served as our Testing data set. We note that the columns "i1_dep_1_place" and "i1_rcf_1_place" are not needed for this analysis because it only represent the unique i.d. of the intermediate airline for the transportation process, as well as the column "i1_legid" represent the i.d. of the customer.</p>

```python
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
```

```python
first_seg_first_leg["total_dlv_p"] = first_seg_first_leg["i1_rcs_p"] + \
first_seg_first_leg["i1_dep_1_p"] + first_seg_first_leg["i1_rcf_1_p"]

first_seg_first_leg["total_dlv_e"] = first_seg_first_leg["i1_rcs_e"] + \
first_seg_first_leg["i1_dep_1_e"] + first_seg_first_leg["i1_rcf_1_e"]

first_seg_first_leg.head()
```

<p>In the bext task, we will use K-means clustering. K-means clustering aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean. We cluster the individuals using K-means for k=2,..,20. We choose the ideal number of clusters.</p>

```python
from __future__ import print_function

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

#Choosing the best number of clusters


max_silouette11=0;
k_number11=1;
for i in range(2,21,1):
    clusterer11 = KMeans(n_clusters=i, random_state=10)
    cluster_labels11= clusterer11.fit_predict(X_std11)
    silhouette_avg11 = silhouette_score(X_std11, cluster_labels11)
    if(max_silouette11<silhouette_avg11):
        max_silouette11=silhouette_avg11
        k_number11=i
print('Max silouette score value ',max_silouette11)
print('Best number of clusters ',k_number11)
```

```python
#function for plotting silouette and clusteres points on the factorial plane
def plot_silouette(n_clusters):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    ax1.set_xlim([-0.1, 1])   
    ax1.set_ylim([0, len(X_std11) + (n_clusters + 1) * 10])
#Initialize the clusterer with n_clusters value and a random generator
    clusterer11 = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels11 = clusterer11.fit_predict(X_std11)

    silhouette_avg11 = silhouette_score(X_std11, cluster_labels11)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg11)
    
    #Compute the silhouette scores for each sample
    sample_silhouette_values11 = silhouette_samples(X_std11, cluster_labels11)
    
    y_lower11 = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values11 = sample_silhouette_values11[cluster_labels11 == i]

        ith_cluster_silhouette_values11.sort()

        size_cluster_i11 = ith_cluster_silhouette_values11.shape[0]
        y_upper11 = y_lower11 + size_cluster_i11

        color = plt.cm.Spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower11, y_upper11),
                          0, ith_cluster_silhouette_values11,facecolor=color, edgecolor=color,
                           alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower11 + 0.5 * size_cluster_i11, str(i))

        # Compute the new y_lower for next plot
        y_lower11 = y_upper11 + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg11, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = plt.cm.Spectral(cluster_labels11.astype(float) / n_clusters)#c=colors,
    ax2.scatter(X_std11[:, 0], X_std11[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                 edgecolor='k',color=colors)

    # Labeling the clusters
    centers11 = clusterer11.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers11[:, 0], centers11[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers11):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

    plt.show()         
          
```

In order to decide which eigenvectors can dropped without losing too much information for the construction of lower-dimensional subspace, we need to inspect the corresponding eigenvalues: The eigenvectors with the lowest eigenvalues bear the least information about the distribution of the data; those are the ones can be dropped.
```python
# Making a list of (eigenvalue, eigenvector) tuples
eig_pairs22 = [(np.abs(eig_vals22[i]), eig_vecs22[:,i]) for i in range(len(eig_vals22))]

# Sorting the (eigenvalue, eigenvector) tuples in descending order
eig_pairs22.sort(reverse=True)

print('Eigenvalues in descending order:')
print(pd.DataFrame(eig_pairs22)[0])
```

```python
#Choosing the best number of clusters
max_silouette22=0;
k_number22=1;
for i in range(2,21,1):
    clusterer22 = KMeans(n_clusters=i, random_state=10)
    cluster_labels22= clusterer22.fit_predict(X_std22)
    silhouette_avg22 = silhouette_score(X_std22, cluster_labels22)
    if(max_silouette22<silhouette_avg22):
        max_silouette22=silhouette_avg22
        k_number22=i
#print('Max silouette score value ',max_silouette22)
print('Best number of clusters ',k_number22)
```

```python
#function for plotting silouette and clusteres points on the factorial plane
def plot_silouette_2(n_clusters):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    ax1.set_xlim([-0.1, 1])   
    ax1.set_ylim([0, len(X_std22) + (n_clusters + 1) * 10])
#Initialize the clusterer with n_clusters value and a random generator
    clusterer22 = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels22 = clusterer22.fit_predict(X_std22)

    silhouette_avg22 = silhouette_score(X_std22, cluster_labels22)
    print("For n_clusters =", n_clusters, "The average silhouette_score is :", silhouette_avg22)
    
    #Compute the silhouette scores for each sample
    sample_silhouette_values22 = silhouette_samples(X_std22, cluster_labels22)
    
    y_lower22 = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values22 = sample_silhouette_values22[cluster_labels22 == i]

        ith_cluster_silhouette_values22.sort()

        size_cluster_i22 = ith_cluster_silhouette_values22.shape[0]
        y_upper22 = y_lower22 + size_cluster_i22

        color = plt.cm.Spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower22, y_upper22),
                          0, ith_cluster_silhouette_values22,facecolor=color, edgecolor=color,
                           alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower22 + 0.5 * size_cluster_i22, str(i))

        # Compute the new y_lower for next plot
        y_lower22 = y_upper22 + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg22, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = plt.cm.Spectral(cluster_labels22.astype(float) / n_clusters)#c=colors,
    ax2.scatter(X_std22[:, 0], X_std22[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                 edgecolor='k',color=colors)

    # Labeling the clusters
    centers22 = clusterer22.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers22[:, 0], centers22[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers22):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

    plt.show()         
```
